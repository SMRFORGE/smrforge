```python
"""SMRForge Setup + Test Data

This notebook does two things:
1) Verifies you can import SMRForge
2) (Optional) Regenerates manual test data under `testing/test_data/`

Nothing here is required for the automated pytest suite.
"""

from __future__ import annotations

import runpy
from pathlib import Path


def find_repo_root(start: Path) -> Path:
    for p in (start, *start.parents):
        if (p / "pytest.ini").exists() and (p / "smrforge").is_dir() and (p / "testing").is_dir():
            return p
    raise RuntimeError("Could not find repo root (expected pytest.ini + smrforge/ + testing/)")


repo = find_repo_root(Path.cwd())
print(f"Repo root: {repo}")

# 1) Import check
import smrforge as smr

print("SMRForge version:", getattr(smr, "__version__", "<unknown>"))

# 2) Optional: regenerate manual test data
script = repo / "testing" / "generate_test_data.py"
print(f"\n(Optional) Run test data generator: {script}")
print("If you don't want to regenerate data, just skip running this cell.")

try:
    runpy.run_path(str(script), run_name="__main__")
except SystemExit as e:
    code = int(e.code) if isinstance(e.code, int) else 0
    print(f"Test data generator exited with code: {e.code}")
    if code != 0:
        raise
```
