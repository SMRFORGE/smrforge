```python
"""Feature: Parameter sweep

Runs: testing/test_04_parameter_sweep.py
"""

from __future__ import annotations

import runpy
from pathlib import Path


def find_repo_root(start: Path) -> Path:
    for p in (start, *start.parents):
        if (p / "pytest.ini").exists() and (p / "smrforge").is_dir() and (p / "testing").is_dir():
            return p
    raise RuntimeError("Could not find repo root (expected pytest.ini + smrforge/ + testing/)")


repo = find_repo_root(Path.cwd())
script = repo / "testing" / "test_04_parameter_sweep.py"
print(f"Running: {script}")

try:
    runpy.run_path(str(script), run_name="__main__")
except SystemExit as e:
    code = int(e.code) if isinstance(e.code, int) else 0
    print(f"Exited with code: {e.code}")
    if code != 0:
        raise
```
