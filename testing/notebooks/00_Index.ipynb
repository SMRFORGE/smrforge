```python
"""SMRForge Feature Notebooks (Index)

These notebooks run the manual feature test scripts under `testing/test_*.py`.
They are small wrappers so the scripts remain the single source of truth.

Suggested flow:
- Run `00_Setup_and_Test_Data.ipynb` (optional)
- Run any feature notebook `01_...` through `13_...`
"""

from __future__ import annotations

from pathlib import Path


def find_repo_root(start: Path) -> Path:
    for p in (start, *start.parents):
        if (p / "pytest.ini").exists() and (p / "smrforge").is_dir() and (p / "testing").is_dir():
            return p
    raise RuntimeError("Could not find repo root (expected pytest.ini + smrforge/ + testing/)")


repo = find_repo_root(Path.cwd())
notebooks_dir = repo / "testing" / "notebooks"

print(f"Repo root: {repo}")
print(f"Notebooks: {notebooks_dir}")
print()

print("Feature notebooks:")
for nb in sorted(notebooks_dir.glob("*.ipynb")):
    if nb.name.startswith("00_"):
        continue
    print(f"- {nb.name}")
print()

print("Manual test scripts:")
for script in sorted((repo / "testing").glob("test_*.py")):
    print(f"- {script.name}")
```
